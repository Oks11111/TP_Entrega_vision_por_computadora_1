```python

import cv2 as cv
import time
import numpy as np
import matplotlib.pyplot as plt

img_color1 = cv.imread(r'C:\Users\54922\GITHUB\vision_computadora_I\Material_TPs\TP1\WP_B.png')
img_color2 = cv.imread(r'C:\Users\54922\GITHUB\vision_computadora_I\Material_TPs\TP1\WP_O.png')
img_color3 = cv.imread(r'C:\Users\54922\GITHUB\vision_computadora_I\Material_TPs\TP1\WP_R.png')

imgHSV1 = cv.cvtColor(img_color1,cv.COLOR_BGR2HSV)
imgHSV2 = cv.cvtColor(img_color2,cv.COLOR_BGR2HSV)
imgHSV3 = cv.cvtColor(img_color3,cv.COLOR_BGR2HSV)


plt.imshow(img_color1[:,:,::-1])
plt.show()
plt.imshow(img_color2[:,:,::-1])
plt.show()
plt.imshow(img_color3[:,:,::-1])
plt.show()
```

```python
#while(True):
 #   start_time = time.time()
for x in [imgHSV1,imgHSV2,imgHSV3]:
    B,G,R = cv.split(x)

    maxr=np.amax(R)
    maxb=np.amax(B)
    maxg=np.amax(G)


    sr=float(255)/maxr
    sg=float(255)/maxg
    sb=float(255)/maxb


    image1_=cv.merge([B,G,R])
    image1 = cv.cvtColor(image1_, cv.COLOR_HSV2RGB)
    plt.imshow(image1[:,:,::-1])
    plt.show()

    image2_=cv.merge([B*sb,G*sg,R*sr])
    image2 = cv.cvtColor(image2_.astype('uint8'), cv.COLOR_HSV2RGB)
    plt.imshow(image2[:,:,::-1])
    plt.show()
```
